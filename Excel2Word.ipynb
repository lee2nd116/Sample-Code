```python
import tkinter as tk
import xlwings as xw
from tkinter import messagebox
from tkinter import filedialog as fd
import os
import win32com.client
from PIL import ImageTk, Image
import datetime


window = tk.Tk()
window.title('Datamap Word Comparison')
window.geometry('410x500')
window.configure(background='LightBlue1')
        

def Word2Excel():
    
    """
    抓 vba 程式,
    並把路徑寫死在 G 槽 CI&I SD Team 裡面
    步驟 :
    1.開啟 Excel
    2.開啟 Word2Excel_Tool.xlsm 活頁簿檔案
    3.執行巨集程式
    4.離開 Excel 
    """
    
    excel = win32com.client.Dispatch("Excel.Application")    
    excel.Workbooks.Open(Filename=r"G:\Shared drives\CI&I SD Team\VBA Tools\A103_datamapXword\Word2Excel_Tool.xlsm")    
    excel.Application.Run("Word2Excel_Tool.xlsm!ImportWordTable")    
    
       
def loading1():
    
    global file1
    file1=fd.askopenfilenames(title='Select Datamap')
    LoadingPath1 = '{}'.format(file1)
    Loading_label1.configure(text=LoadingPath1)

    
def loading2():
    
    global file2
    file2=fd.askopenfilenames(title='Select _ExcelxWord')
    LoadingPath2 = '{}'.format(file2)
    Loading_label2.configure(text=LoadingPath2)
    
       
def Comparison():
  
    # 開始測量
    start = datetime.datetime.now()    

    wb1 = xw.Book('%s' % (file1))
    wb2 = xw.Book('%s' % (file2))

    datamap_sheet = wb1.sheets[0]
    ExcelxWord_sheet = wb2.sheets[0]
    
    datamap_sheet_last_row = datamap_sheet.range("A1").end("down").row
    ExcelxWord_sheet_last_row = ExcelxWord_sheet.range("A1048576").end("up").row
    
    GridQSet = []
    GridrLabelSet = []
    GridLabelSet = []
    SMTotalSet = []
    OpenTotalSet = []
    
    
    ## 一行一行把 _ExcelxWord 轉成2維陣列    
    for i in range(ExcelxWord_sheet_last_row+1):
        
        ## Grid         
        if (ExcelxWord_sheet.range(f"A{i+1}").api.Font.Color > 0):    
            
            Q1name = ExcelxWord_sheet.range(f"A{i+1}").value 
            j = 0
            o = 1
            label = 1
        
        
            ## While is None一定要放在條件式前面
            while 1:
                                
                Q1name = ExcelxWord_sheet.range(f"A{i-j+1}").value
                rlabel = ExcelxWord_sheet.range(f"A{i+1}").value
                nPos=rlabel.find(')')
                rlabel = rlabel[nPos+1:]  
            
                j+=1
                
                if (Q1name is None):
                    Q1name = ExcelxWord_sheet.range(f"A{i-j+3}").value
                    
                    if (Q1name[0] == 'Q'):                   
                        break
                    else:
                        continue
                    
                                        
            while 1:

                label = ExcelxWord_sheet.range(f"B{i-j+3}").value
                label = str(label)
                
                if (label is not None) and (label[0] == "Q"):
                     label = ExcelxWord_sheet.range(f"A{i-j+3}").offset(1,o).value
                else:
                     label = ExcelxWord_sheet.range(f"A{i-j+3}").offset(0,o).value
                                       
                o+=1 
                
                if (label is None):
                    
                    break
                    
                GridLabelSet.append(label)
                                                   
            GridQSet.append(Q1name)        
            GridrLabelSet.append(rlabel)        
         
        GridQSet = list(set(GridQSet))
        GridrLabelSet = list(set(GridrLabelSet))
        GridLabelSet = list(set(GridLabelSet))
        
        ## Single/Multi
        if (((ExcelxWord_sheet.range(f"A{i+1}").value) is not None) and
        (ExcelxWord_sheet.range(f"A{i+1}").value[0]=='Q') and
        (isinstance(ExcelxWord_sheet.range(f"B{i+1}").value, str)) and  
        (isinstance(ExcelxWord_sheet.range(f"C{i+1}").value, float))):
        
            SMSet = []
            Q2name = ExcelxWord_sheet.range(f"A{i+1}").value
            Q2code = ExcelxWord_sheet.range(f"C{i+1}").value
            Q2label = ExcelxWord_sheet.range(f"B{i+1}").value
            
            Q2code = int(Q2code)            
            
            SMSet.append(Q2name)      ##0
            SMSet.append(Q2code)      ##1
            SMSet.append(Q2label)     ##2
            
            SMTotalSet.append(SMSet)
            

        ## Opentext
        if (((ExcelxWord_sheet.range(f"A{i+1}").value) is not None) and
        (ExcelxWord_sheet.range(f"A{i+1}").value[0]=='Q') and
        ('____________________' in str(ExcelxWord_sheet.range(f"B{i+1}").value)) and  
        ((ExcelxWord_sheet.range(f"C{i+1}").value) is None)):
            
            Q3name = ExcelxWord_sheet.range(f"A{i+1}").value
            OpenTotalSet.append(Q3name)
            OpenTotalSet.append(Q3name+'R1')
            OpenTotalSet.append(Q3name+'R2')
            OpenTotalSet.append(Q3name+'R3')
            OpenTotalSet.append(Q3name+'R4')
            OpenTotalSet.append(Q3name+'R5')
            OpenTotalSet = set(OpenTotalSet)
            OpenTotalSet = list(OpenTotalSet)
            

        ## Vertical Grid
        if (((ExcelxWord_sheet.range(f"C{i+1}").value) is not None) and
        (str(ExcelxWord_sheet.range(f"C{i+1}").value)[0]=='Q')):
            
            Q4name = ExcelxWord_sheet.range(f"C{i+1}").value[:-1]
            Q4namea = Q4name + 'a' 
            Q4nameb = Q4name + 'b'
            Q4namec = Q4name + 'c'
            Q4named = Q4name + 'd'
            Q4namee = Q4name + 'e'
           
            k = 1
                        
            while 1:
                
                code1 = ExcelxWord_sheet.range(f"C{i+k+1}").value
                k+=1
                
                if code1 == 1 :
                    break
                    
            m = 0
            
            while 1:
                
                VGSeta = []
                VGSetb = []
                VGSetc = []
                VGSetd = []
                VGSete = []                
                
                if (ExcelxWord_sheet.range(f"A{i+k+m}").value is None) or (ExcelxWord_sheet.range(f"A{i+k+m}").value[0]=='Q'):
                    break
                
                if (ExcelxWord_sheet.range(f"A{i+k+m}").value is not None) and (ExcelxWord_sheet.range(f"B{i+k+m}").value is not None): 
                    Q4label = ExcelxWord_sheet.range(f"A{i+k+m}").value
                    Q4code = ExcelxWord_sheet.range(f"B{i+k+m}").value
                        
                        
                m+=1 
                
                VGSeta.append(Q4namea)  
                VGSeta.append(Q4code)   
                VGSeta.append(Q4label)   
                SMTotalSet.append(VGSeta)
                
                VGSetb.append(Q4nameb)                 
                VGSetb.append(Q4code)      
                VGSetb.append(Q4label)     
                SMTotalSet.append(VGSetb)  
                                           
                VGSetc.append(Q4namec)     
                VGSetc.append(Q4code)      
                VGSetc.append(Q4label)     
                SMTotalSet.append(VGSetc)  
                                           
                VGSetd.append(Q4named)     
                VGSetd.append(Q4code)      
                VGSetd.append(Q4label)     
                SMTotalSet.append(VGSetd)  
                                           
                VGSete.append(Q4namee)     
                VGSete.append(Q4code)      
                VGSete.append(Q4label)     
                SMTotalSet.append(VGSete)  
                
#     print(OpenTotalSet)
#     print(len(OpenTotalSet))
#     print()
#     print(SMTotalSet)
#     print(len(SMTotalSet))
#     print()
#     print(GridQSet)
#     print(len(GridQSet))
#     print() 
#     print(GridrLabelSet)
#     print(len(GridrLabelSet))
#     print() 
#     print(GridLabelSet)
#     print(len(GridLabelSet))
#     print()     
    
                              
    # 正式開始比對
    for i in range(datamap_sheet_last_row):
        
        Qdname = datamap_sheet.range(f"A{i+1}").value
        Vdname = datamap_sheet.range(f"B{i+1}").value
        Qdcode = datamap_sheet.range(f"E{i+1}").value
        Qdlabel = datamap_sheet.range(f"G{i+1}").value
        Adlabel = datamap_sheet.range(f"H{i+1}").value
        Qdtype = datamap_sheet.range(f"I{i+1}").value
        
        if (Qdcode == "1") and (Qdtype == "multi"):
        
            nPos1=Vdname.find("_")
            Qdcode = Vdname[nPos1+1:]
            Qdcode = float(Qdcode)

                 
        CompareSet = []
        
        #全開 ---> 標紅色
        if (Qdname[0] == "Q") and (Qdname == Vdname) and (Qdtype == "opentext") and (Qdname not in OpenTotalSet): 
                         
            datamap_sheet.range(f"A{i+1}").color = (250,99,71)

            
        #Single/Multi ---> 標黃色
        if (Qdname[0] == "Q") and (Qdtype == "single" or Qdtype == "multi"):    
            
            CompareSet.append(Qdname)
            CompareSet.append(Qdcode)
            CompareSet.append(Adlabel)

            if (CompareSet not in SMTotalSet):
                
                datamap_sheet.range(f"A{i+1}").color = (240,230,140) 
                datamap_sheet.range(f"E{i+1}").color = (240,230,140) 
                datamap_sheet.range(f"H{i+1}").color = (240,230,140) 
                
   
        ##Grid ---> 標綠色
        if ((Qdname[0] == "Q") and (Qdtype == "grid") and  
            (not ((Qdname in GridQSet) and (Qdlabel in GridrLabelSet) and (Adlabel in GridLabelSet)))):
            
            datamap_sheet.range(f"A{i+1}").color = (189,252,201)
            datamap_sheet.range(f"B{i+1}").color = (189,252,201)
            datamap_sheet.range(f"G{i+1}").color = (189,252,201)
            datamap_sheet.range(f"H{i+1}").color = (189,252,201)        

  
    end = datetime.datetime.now()
    duration = end - start
    result = 'Run Time：{}'.format(duration)
    result_label.configure(text=result)            
    
    messagebox.showinfo("Message", "Comparison completed")            
   
    
"""
GUI 介面
"""      

# STG Datamap Comparison Title    
header_label = tk.Label(window, text='Comparison\n Datamap ↔ Word ', width=24, height=2, bg='gray80', fg='LightSkyBlue4')
header_label.config(font='helvetica 18 bold')
header_label.place(x = 20, y = 20)


# Insert Nielsen Image
img = tk.PhotoImage(file=r"G:\Shared drives\CI&I SD Team\VBA Tools\Development tools (Axxx)\A103_DatamapxWord\Nielsen.png") #圖片位置
Nielsen = tk.Label(window, image = img, bg = 'LightBlue1', width=315, height=140)
Nielsen.place(x = 115, y = 375)


# Function Button
Word2Excel_btn = tk.Button(window, activebackground='azure3', text='Word to Excel', command=Word2Excel)
Word2Excel_btn.place(x = 25, y = 120)

notice_label = tk.Label(window, text='After execution, please press "Yes"', width=28, height=1, bg='LightBlue1', fg='dark green')
notice_label.config(font='helvetica 10 bold')
notice_label.place(x = 130, y = 120)


# Load Datamap file Button
Loading_btn1 = tk.Button(window, activebackground='azure3', text='Load Datamap file', command=loading1)
Loading_btn1.place(x = 25, y = 170)


# Print Load Datamap file Path
Loading_label1 = tk.Label(window,bg = 'LightBlue1')
Loading_label1.place(x = 25, y = 200)


# Load _ExcelxWord file Button
Loading_btn2 = tk.Button(window, activebackground='azure3', text='Load _ExcelxWord file', command=loading2)
Loading_btn2.place(x = 25, y = 230)


# Print _ExcelxWord Datamap file Path
Loading_label2 = tk.Label(window,bg = 'LightBlue1')
Loading_label2.place(x = 25, y = 260)


# Function Button
Comparison_btn = tk.Button(window, activebackground='azure3', text='Compare', command=Comparison)
Comparison_btn.place(x = 25, y = 300)


# Print Runtime result
result_label = tk.Label(window,bg = 'LightBlue1',fg = 'red2')
notice_label.config(font='helvetica 10 bold')
result_label.place(x = 120, y = 300)


# Result Notice
notice_label = tk.Label(window, text='**Notice the colored cells in Datamap.xlsx', width=33, height=1, bg='LightBlue1', fg='dark green')
notice_label.config(font='helvetica 10 bold')
notice_label.place(x = 23, y = 340)
notice_label = tk.Label(window, text='*Single/Multi ---> Yellow ', width=19, height=1, bg='LightBlue1', fg='gold2')
notice_label.config(font='helvetica 9 bold')
notice_label.place(x = 24, y = 370)
notice_label = tk.Label(window, text='*Full-OE ---> Red', width=13, height=1, bg='LightBlue1', fg='red2')
notice_label.config(font='helvetica 9 bold')
notice_label.place(x = 23, y = 390)
notice_label = tk.Label(window, text='*Grid ---> Green ', width=13, height=1, bg='LightBlue1', fg='green3')
notice_label.config(font='helvetica 9 bold')
notice_label.place(x = 25, y = 410)


# Quit Button
tk.Button(window, text="Cancel", activebackground='LightSkyBlue3',width = 10, command=window.destroy).place(x = 30, y = 450)


window.mainloop()
```
